<center><h2>Analyze the dataset for the IELTS Speaking Part 2 task</h2></center>

```python
import json
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

<h3>Open file all_data.json</h3>

```python
with open('all_data.json') as file:
    all_datasets = json.load(file)

print(f"Length : {len(all_datasets)}, Type : {type(all_datasets)}")
all_datasets[:2]
```

```python
all_datasets_df = pd.DataFrame(all_datasets)
all_datasets_df.head(5)
```

```python
question_unique = all_datasets_df["question"].value_counts()
question_unique
```

```python
with open('question_file_test_output_json.json') as file:
    questions_df = json.load(file)

print(f"Length : {len(questions_df)}, Type : {type(questions_df)}")
questions_df[:2]
```

```python
list_question = [i["question"] for i in questions_df]
list_question
```

```python
list_count = []
for question in list_question:
    try:
        list_count.append(question_unique[question])
    except:
        list_count.append(0)
list_count
```

```python
question_df = pd.DataFrame({"question": list_question, "duplicate": list_count})
question_df
```

```python
question_df.to_json(r'question_file_test_output_json.json', orient="records")
```
